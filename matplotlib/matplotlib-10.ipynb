```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import datetime
from collections import Counter
```

```python
###time series graph

plt.style.use('seaborn-v0_8')
ds = pd.read_csv("data10.csv")
ds.info()

age=ds["Age"]
all_dev_salaries = ds["All_Devs"]
py_salaries = ds["Python"]
js_salaries = ds["JavaScript"]

plt.plot(age,py_salaries, label="Python")
plt.plot(age,js_salaries, label="Javascript")

plt.plot(age,all_dev_salaries, label="All Devs" )

plt.legend(loc="upper left")
plt.title("Dev Salaries")
plt.tight_layout()
#plt.savefig("matplot-graph-8.png")
plt.show()
```
