```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from collections import Counter
```

```python
###Sample pie chart

plt.style.use('ggplot')
slices = [60,40]
labels=["Sixty","Forty"]
colors=["yellow","green"]
plt.pie(slices, labels=labels,colors=colors,wedgeprops={'edgecolor':"black"})
plt.title("Pie Chart")
plt.legend()
plt.tight_layout()
plt.savefig("")
plt.show()
```

```python
# Language Popularity
slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
explode = [0,0,0,0.1,0]
plt.style.use('ggplot')
plt.pie(slices, labels=labels,explode = explode, shadow=True,startangle=90,autopct="%.2f", wedgeprops={'edgecolor':"black"})
plt.title("Pie Chart")
plt.tight_layout()
plt.savefig("")
plt.show()
```
