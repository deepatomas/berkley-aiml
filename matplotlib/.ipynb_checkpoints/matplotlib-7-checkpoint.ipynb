```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from collections import Counter
```

```python
###histgram

plt.style.use('seaborn-v0_8')
ds = pd.read_csv("data7.csv")
ds.info()
view_count=ds["view_count"]
likes=ds["likes"]
ratio = ds['ratio']
plt.scatter(x=view_count, y=likes,c=ratio, cmap="winter", edgecolor="black", linewidth=1, alpha=0.75,label="likes")
plt.xscale('log')
plt.yscale('log')
cbar = plt.colorbar()
cbar.set_label="Likes/Dislike ratio"
plt.title("Trending youtube videos")
plt.xlabel("views")
plt.ylabel("likes")
plt.legend(loc="upper left")
plt.tight_layout()
plt.savefig("matplot-graph-7.png")
plt.show()
```
