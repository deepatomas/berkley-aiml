```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from collections import Counter
```

```python
###stack chart

plt.style.use('fivethirtyeight')
ds = pd.read_csv("data5.csv")
ds.info()
age=ds["Age"]
dev_salaries=ds["All_Devs"]
py_salaries=ds["Python"]
js_salaries=ds["JavaScript"]
overall_median = 57287
plt.plot(age,py_salaries ,label="Python Devs")
plt.plot(age,dev_salaries, color="k", linestyle="--",label="All Devs")
plt.fill_between(age,py_salaries,dev_salaries,where =(py_salaries > dev_salaries), label="Above all devs")
plt.fill_between(age,py_salaries,dev_salaries,where =(py_salaries < dev_salaries), label="Below all dev")

plt.title("Fill plt")
plt.legend(loc="upper left")
plt.tight_layout()
plt.savefig("matplot-graph-5.png")
plt.show()
```
