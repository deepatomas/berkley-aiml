```python
###generate horizontal barchart with csv file

languages =[]
popularity = []
plt.style.use('ggplot')
plt.figure(figsize=(10,6))
ds = pd.read_csv("data.csv")
ds.info()
ds.head()

#loop through the daraframe to create a dictionary of langauge counters counting 
c= Counter([])
for index in range(len(ds)):
    c.update(ds.loc[index, "LanguagesWorkedWith"].split(";"))

for item in c.most_common(15):
    languages.append(item[0])
    popularity.append(item[1])

languages.reverse()
popularity.reverse()


plt.barh(languages,popularity , label="Popular tools")
plt.ylabel("Programming languages")
plt.xlabel("Number of users")
plt.legend()
plt.savefig("matplot-graph-2-2.png")
plt.show()
```
